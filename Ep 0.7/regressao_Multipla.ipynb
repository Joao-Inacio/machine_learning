# **Regressão Polinomial - Machine Learning 07**

### Importação

```python
# Libs
import numpy as np
import pandas as pd
```

### Base de dados

```python
# Função de carregamento de dados
def carregarDados(arquivo):
    ds = pd.read_csv(arquivo, sep=';')
    # Amostra
    X = ds.iloc[:,:-1].values
    # Variável independente 
    y = ds.iloc[:,-1].values
    return X, y, ds
```

### Gráfico

```python
# Função de Plotar gráfico
def plotarGraphPol(X_Points, y_Points, X_Line, y_Line):
    import matplotlib.pyplot as plt
    plt.scatter(X_Points, y_Points, color='red')
    plt.plot(X_Line, y_Line, color='blue')
    plt.title("Comparando pontos reais com a reta produzida pela regressão polinomial")
    plt.xlabel("Experiéncia em anos")
    plt.ylabel("Salário")
    plt.show()
```

```python
def plotarGraphLinear(X, y, linearRegressor):
    import matplotlib.pyplot as plt

    plt.scatter(X, y, color = 'red') #plot real y points
    plt.plot(X, linearRegressor.predict(X), color = 'blue') #plot predicted points in line
    plt.title("Comparando pontos reais com a reta produzida pela regressão linear")
    plt.xlabel("Experiência em anos")
    plt.ylabel("Salário")
    plt.show()
```

### Regressão Linear

```python
# Função de Regressão Linear
def modeloRegressaoLinear(X, y):
    from sklearn.linear_model import LinearRegression
    regressor = LinearRegression()
    regressor.fit(X, y)
    return regressor
```

```python
# Função de regressão linear
def regressaoLinearPolinomial(X, y, d):
    from sklearn.preprocessing import PolynomialFeatures
    polynomial_Features = PolynomialFeatures(degree = d)
    X_Polynomial = polynomial_Features.fit_transform(X)
    
    from sklearn.linear_model import LinearRegression
    polLinear_Regression = LinearRegression()
    polLinear_Regression.fit(X_Polynomial, y)

    return X_Polynomial, polLinear_Regression
```

### Rodando Todas as Funções

```python
# Função que roda todas as funções
def rodarRegressao(dados):
    X, y, csv = carregarDados(dados)

    linearRegressor = modeloRegressaoLinear(X, y)
    
    XPoly, polyLinearRegressor = regressaoLinearPolinomial(X, y, 4)

    plotarGraphLinear(X, y, linearRegressor)
    plotarGraphPol(X, y, X, polyLinearRegressor.predict(XPoly))
```

### Resultado Final

```python
# # Rodando a função 
rodarRegressao('../data/salary.csv')
```
