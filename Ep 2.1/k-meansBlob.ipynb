### Bibliotecas

```python
# Libs
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
from sklearn.datasets import make_blobs
```

### Criando os centroides 

```python
X, y = make_blobs(n_samples=30, centers=3, random_state=0)
plt.scatter(X[:, 0], X[:, 1]);
```

```python
# Sorteando o primeiro centroide 
k = 3
num_samples, num_dimensions = X.shape

# escolhendo o centroide
np.random.seed(0)
centroids = np.zeros((k, num_dimensions))
rand_index = np.random.choice(num_samples)
centroids[0] = X[rand_index]

# Visualizando
plt.scatter(X[:, 0], X[:, 1], color='blue')
plt.scatter(centroids[0][0], centroids[0][1], color='red')
plt.show()
```

```python
# computa as dist√¢ncias do primeiro centr√≥ide para todas as amostras
distances_to_centroids = pairwise_distances(X, [centroids[0]], metric='euclidean').flatten()
distances_to_centroids
```

```python
# Computando os dois centroides que falta
for i in range(1, k):
    prob = distances_to_centroids ** 2
    rand_index = np.random.choice(num_samples, size=1, p=prob / np.sum(prob))
    print(f'Probabilidade de cada cluster ser escolhido: {prob / np.sum(prob)}\n')
    print(f'Cluster escolhido: {rand_index}\n')
    centroids[i] = X[rand_index]

    distances_new = pairwise_distances(X, [centroids[i]], metric='euclidean').flatten()
    print(f'Dist√¢ncia de cada amostra ao centr√≥ide {i - i}: {distances_to_centroids}\n')
    print(f'Dist√¢ncia de cada amostra ao centr√≥ide {i}: {distances_new}\n')

    distances_to_centroids = np.min(np.vstack((distances_to_centroids, distances_new)), axis = 0)
    print(f'Dist√¢ncias atualizadas: {distances_to_centroids}\n')

    plt.scatter(X[:, 0], X[:, 1], color='blue')
    for j in range(0, i + 1):
        plt.scatter(centroids[j,0], centroids[j,1], color='red')
    plt.show()
```
