# Preenchendo Dados Faltantes

### Bibliotecas

```python
# Libs
import numpy as np
import pandas as pd
```

### Base de dados

```python
# Dados
ds = pd.read_csv('../data/svbr.csv', sep=';')
ds.head()
```

### Separando os valores

```python
# Valores na variável x
X = ds.iloc[:,:].values
X
```

### Substituindo os valores

```python
# Lib
from sklearn.impute import SimpleImputer
```

```python
# Preenchendo os valores
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer = imputer.fit(X[:, 1:3])
X = imputer.transform(X[:, 1:3]).astype(str)
X = np.insert(X, 0, ds.iloc[:, 0].values, axis=1)
X
```

### Transformando de volta em DataFrame

```python
# DataFrame
ds = pd.DataFrame(X, columns=['Canal', 'Inscritos' ,'Visualizações'])
ds
```
